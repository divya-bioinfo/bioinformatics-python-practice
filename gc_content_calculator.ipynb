```python
dna = "ATGCGTAGCTAAGCT"
A = dna.count("A")
T = dna.count("T")
G = dna.count("G")
C = dna.count("C")

length = len(dna)

gc_content = (G + C) / length * 100

print("DNA sequence:", dna)
print("A:", A)
print("T:", T)
print("G:", G)
print("C:", C)
print("GC Content:", round(gc_content, 2), "%")
```
